```python
 import pandas as pd
 import numpy as np
 import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('drdo dataset.csv')
```

```python
dataset.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(dataset.iloc[:,[0]],dataset.iloc[:,[1]],test_size=0.2,random_state=2)
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
X_train.shape
```

```python
y_class = y_train.round().astype(int)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=13)
knn.fit(X_train,y_class)
```

```python
y_tests= y_test.round().astype(int)
```

```python
from sklearn.metrics import accuracy_score
y_pred=knn.predict(X_test)
accuracy_score(y_tests,y_pred)
```

```python
score=[]
for i in range (1,15):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_class)
    y_pred=knn.predict(X_test)
    score.append(accuracy_score(y_tests,y_pred))
plt.plot(range(1,15),score)
```
